```python
from text_cleanisation_NLP import text_to_words
import collections
import math
```

```python
f = open('restaurants_corpus.txt' , 'r' , encoding='utf-8-sig')
text = f.read()
```

```python
resto = text_to_words(text)
```

```python
f.close()
f2 = open('contrast_corpus.txt' , 'r' , encoding='utf-8-sig')
text2 = f2.read()
contrast = text_to_words(text2)
```

```python
def compute_tf(text):
    tf_text = collections.Counter(text)
    for i in tf_text:
        tf_text[i] = tf_text[i]/float(len(text))
    return tf_text
```

```python
corpus = []
corpus.append(resto)
corpus.append(contrast)
```

```python
def compute_idf(word, corpus):
    return math.log10(len(corpus)/sum([1.0 for i in corpus if word in i]))
```

```python
def compute_tfidf(corpus):
    documents_list = []
    for text in corpus:
        tf_idf_dictionary = {}
        computed_tf = compute_tf(text)
        for word in computed_tf:
            tf_idf_dictionary[word] = computed_tf[word] * compute_idf(word, corpus)
        documents_list.append(tf_idf_dictionary)
    return documents_list
```

```python
tf_idf = compute_tfidf(corpus)
```

```python
tf_idf_resto = tf_idf[0]
top = {}
for i in tf_idf_resto:
    if tf_idf_resto[i] != 0.0:
        if tf_idf_resto[i] not in top.keys():
            top[tf_idf_resto[i]] = []
            top[tf_idf_resto[i]].append(i)
        else:
            top[tf_idf_resto[i]].append(i)

```

```python
for w in sorted(top, reverse = True):
    print ('score = ' + str(w))
    for i in top[w]:
        print (i)
    print('------------------')
```
